```python
with open('VHDL/input.txt', 'r') as f:
  input_converted = []
  j = 0
  for line in f:
    line = line.strip()

    if j == 0:
      input_converted.append([0] * (len(line) + 2))

    j += 1
    input_converted.append([0])

    for i in range(len(line)):
      if line[i] == '.':
        input_converted[j].append(0)
      else:
        input_converted[j].append(1)


    input_converted[j].append(0)

input_converted.append([])

for i in range(len(line) + 2):
  input_converted[j + 1].append(0)

valid = 0


output = []

while True:
  indices = []
  validprev = valid

  for i in range(1 ,len(input_converted) - 1):
    output.append([])
    for j in range(1 ,len(input_converted[0]) - 1):
      sum = input_converted[i - 1][j - 1]
      sum += input_converted[i][j - 1]
      sum += input_converted[i + 1][j - 1]
      sum += input_converted[i - 1][j]
      sum += input_converted[i + 1][j]
      sum += input_converted[i - 1][j + 1]
      sum += input_converted[i][j + 1]
      sum += input_converted[i + 1][j + 1]
      if sum < 4 and input_converted[i][j] == 1:
        valid += 1
        indices.append([i,j])

  for i in range(len(indices)):
    input_converted[indices[i][0]][indices[i][1]] = 0
  # print(input_converted)
  if valid == validprev:
    break



print(valid)

```
