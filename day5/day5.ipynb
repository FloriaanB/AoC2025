```python
def remove_overlap(ranges):
    result = []
    current_start = -1
    current_stop = -1

    for start, stop in sorted(ranges):
        if start > current_stop + 1:
            # this segment starts after the last segment stops
            # just add a new segment
            result.append( (start, stop) )
            current_start, current_stop = start, stop
        elif stop > current_stop:
            # segments overlap, replace
            result[-1] = (current_start, stop)
            # current_start already guaranteed to be lower
            current_stop = max(current_stop, stop)


    return result
```

```python
def part2():
  with open('VHDL/input.txt', 'r') as f:
    ranges = []
    for line in f:
      line = line.strip()
      if len(line) > 0 :
        for i in range(len(line)):
          if line[i] == "-":
            ranges.append([int(line[:i]),int(line[i+1:])])
            break

  total_fresh = 0

  ranges_comp = remove_overlap(ranges)

  print(ranges_comp)

  for j in range(len(ranges_comp)):
    total_fresh += ranges_comp[j][1] - ranges_comp[j][0] + 1

  return total_fresh

print(part2())
```
