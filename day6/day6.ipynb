```python
with open('input.txt', 'r') as f:
  # print(f[0])
  linefull = []
  for line in f:
    line = line.strip()
    # line = line.strip(" ")
    line = line.split(" ")

    # print(line)
    # line = line.remove('')

    linenew = []


    for i in range(len(line)):
      if len(line[i]) != 0:
        # print(line)

        linenew.append(line[i])

    linefull.append(linenew)




linefull = list(zip(*linefull))

sum = 0
for i in range(len(linefull)):

  if linefull[i][-1] == "*":
    result = 1
    # print(len(linefull[0]))
    for j in range(len(linefull[0]) - 1):
      result *= int(linefull[i][j])
  else:
    result = 0
    # print(len(linefull[0]))
    for j in range(len(linefull[0]) - 1):
      result += int(linefull[i][j])


  sum += result

print(sum)
```

```python

with open('input.txt', 'r') as f:
  # print(f[0])
  linefull = []
  max_len = 0
  for line in f:
    line = line.strip("\n")

    # line = line.split(" ")

    parts = [line[i:i+1] for i in range(0, len(line), 1)]
    linefull.append(parts)

    if len(parts) > max_len:
      max_len = len(parts)


for i in range(len(linefull)):
  while len(linefull[i]) < max_len:
    linefull[i].append(' ')


linefull = list(zip(*linefull))

i_save = 0
result = 1

sum = 0

for i in range(len(linefull)):
  j = 0
  value = 0
  for j in range(len(linefull[i])- 1):
    if linefull[i][j] != ' ':
      value = value * 10 + int(linefull[i][j])

  if value == 0:
    i_save = i + 1
    sum += result
    # print(sum)

    if linefull[i_save][-1] == "*":
      result = 1
    else:
      result = 0

  elif linefull[i_save][-1] == "*":
    result *= value
  else:
    result += value


print(result)
sum += result

print(sum)
```
