```python

def part1():
  with open('input.txt', 'r') as f:
    score = 50
    code = 0
    for line in f:

      score_add = int(line[1:])

      if score_add > 99:
        score_add = score_add % 100



      if line[0] == "L":
        score = score - score_add
      elif line[0] == "R":
        score = score + score_add

      if score < 0:
        score = 100 + score
      elif score >= 100:
        score = score - 100
      # print(score)

      if score == 0:
        code += 1

    return code

print(part1())
```

```python

def part2():
  with open('VHDL/test.txt', 'r') as f:
    score = 50
    code = 0
    for line in f:

      score_add = int(line[1:])

      if score_add > 99:
        score_add_norm = score_add % 100
        hunderds = (score_add - score_add_norm) / 100
        code += hunderds

        # print(hunderds)

        score_add = score_add_norm

      prev_score = score

      if line[0] == "L":
        score = score - score_add
      elif line[0] == "R":
        score = score + score_add

      print(score)
      if score < 0:
        score = 100 + score
        if prev_score != 0:
          code += 1
      elif score > 100:
        score = score - 100
        if prev_score != 0:
          code += 1
      elif score == 100:
        score = 0

      if score == 0:
        code += 1

      print(score, code)

    return code

print(part2())
```
